## Gráfico de línea

El gráfico de línea es una de las maneras más simples y usadas dentro de las técnicas de visualización de datos. Un gráfico de línea muestra información como una serie de puntos o marcas conectadas por líneas rectas. Se puede ajustar el tamaño, color y otros elementos de las líneas y marcas para poder verlo más claramente 

Supongamos que acá tenemos una lista con la cantidad de manzanas producidas (toneladas por hectárea) a lo largo de unos seis años, en un país desconocido. 

```python
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

Podemos visualizar como cambia la producción de manzanas a través del tiempo usando un gráfico de línea. Para dibujarlo usamos la función `%plt.plot` 

```python
plt.plot(yield_apples)
```

Observamos que devuelve un gráfico de línea junto con un línea del plot dibujado. Para que solo devuelva un gráfico puedo usar un `;`.

```python
plt.plot(yield_apples);
```

Ahora vamos a mejorar el gráfico paso a paso!

## Personalizando el eje x
En el gráfico mostrado, el eje x corresponde a una lista de elementos que va del 0 a 5. Sería más útil si encontraramos en el eje x los años en los que estamos ploteando. Esto se puede hacer pasandole dos argumentos a `plt.plot`.

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
```

```python
plt.plot(years, yield_apples); 
```

## Nombres de los ejes.

Podemos agregar etiquetas con los nombres de los ejes usando los métodos `plt.xlabel` y `plt.ylabel`.

```python
plt.plot(years, yield_apples);
plt.xlabel("Year");
plt.ylabel("Yield (ton per hectare)");
```

## Ploteando múltiples lieas

Podemos invocar la función `plt.plot` una vez para cada línea para graficar múltiples líneas en el mismo gráfico. Ahora vamos a comparar la producción de manzanas vs las de naranja.

```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.9375, 0.9372, 0.939]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.898, 0.9, 0.896, ]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

## Cambiar el título y leyenda

Para diferenciar entre múltiples líneas, podemos incluir una leyenda dentro del gráfico usando la función `plt.legend`. También podemos poner un título al gráfico usando la función `plt.title`. 

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

### Line Markers

También se pueden mostrar una marquita para cada punto en los datos en cada linea, usando un argumento de marker en el comando plt.plot. Matplotlib provee diferentes marquitas: circulos, cruces, cuadrados, diamantes, etc. Se puede ver la lista completa de marcas en: https://matplotlib.org/3.1.1/api/markers_api.html .

```python
plt.plot(years, apples, marker='o')
plt.plot(years, oranges, marker='x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges'])
```

## Estilos en lineas y marcas

La funcion plt.plot soporta varios argumentos para cambiar el estilo de las lineas y marquitas. 
Algunos son: 
- *color or c*: Setea el color de la línea
- *linestyle or ls*: Permite elegir entre una linea sólida o "rota". 
- *linewidth or lw*: Setea el ancho de la linea
- *markersize or ms*: Setea el tamaño de la marquita
- *markeredge color or mec*: Setea el color del borde para las marquitas
- *markeredge width or mew*: Setea el tamaño del borde para las marquitas
- *markerfacecolor or mfc*: Setea el color de llenado de las marquitas
- *alpha*: Opacidad del gráfico. 

La documentación completa para plt.plot se puede ver en: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot 


```python
plt.plot(years, apples, marker='s', c='b', ls='-', lw=2, ms=8, mew=2, mec='navy')
plt.plot(years, oranges, marker='o', c='r', ls='--', lw=3, ms=10, alpha=.5)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges']);
```

Se puede especificar más fácil el tamaño de la marquita, el estilo de linea y color con el argumento fmt, que se puede proveer como el tercer argumento de plt.plot: 

`fmt = '[marker][line][color]'`



```python
plt.plot(years, apples, 's-b')
plt.plot(years, oranges, 'o--r')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Oranges'])
```

Si no se especifica un estilo de linea en fmt, se dibuja solo las marquitas. 

```python
plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

## Cambiando el tamaño de la figura

Se puede usar la funcion `plt.figure`para cambiar el tamaño de la figura. 


```python
plt.figure(figsize=(12, 6))

plt.plot(years, oranges, 'or')
plt.title("Yield of Oranges (tons per hectare)");
```

## Gráfico de dispersión (scatter)

En un gráfico de dispersión, el valor de las dos variables está ploteado como puntos en una grilla bi dimensional. Adicionalmente, se puede usar una tercervariable que determine el tamaño o o color de los puntos. Veamos con un ejemplo: 

El Iris dataset provee unas medidas de muestra de sépalos y pétalos para tres especies distintas de flores. Este dataset está incluido en la library de Seaborn y puede ser cargado como un df de Pandas. 

```python
# Cargo los datos en un dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

Vamos a tratar de visualizar las relaciones entre el tamaño y el ancho de un sépalo. Nuestro primer instinto puede ser que creemos un gráfico de línea usando `plt.plot`.

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width)
```

EL output no es muy informativo ya que hay muchas combinaciones de las dos propiedades dentro del dataset. No parece haber una simple relación entre ellos. 

Podemos usar un gráfico de dispersión para analizar cómo varian el largo y ancho de los sépalos usando la de `scatterplot` dentro del módulo de seaborn (imported as sns).

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width);
```

## Agregando matices
Notemos cómo los puntos en el gráfico visto arriba pueden formar distintos clusters con algunos outliers. Podemos colorear usando como parámetro a la especie de la flor. Podemos también hacer los puntos más grandes, usando el argumento s. 

```python
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100);
```

Agregar matices hace que el gráfico se vea más informativo. Podemos decir inmediatamente que las flores de setosa tienen un menor tamaño de sépalo y un mayor ancho. En el contraste, lo opuesto son las flores de Virginica. 

## Histogramas

Un histograma representa la distribución de una variable creando bins (intervalos) en un rango de valores y mostrando en la linea vertical barras que indican el numero de observaciones en cada bin. Por ejemplo, vamos a visualizar los valores del ancho del sépalo en el dataset de flores. Podemos usar la función `plt.hist`para crear un histograma. 

```python
# Load data into a Pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df.sepal_width
```

```python
plt.title("Distribution of Sepal Width")
plt.hist(flowers_df.sepal_width);
```

Podemos ver inmediatamente que los anchos de los sépalos caen en un rango entre los 2.0 y 4.5 y que alrededor de 35 valores están ene el rango de 2.9 - 3.1, lo que se ve como el bin más popular. 

## Ajustando el número y tamaño de bins
Podemos controlar el número de bins o el tamaño de cada uno usando como argumento los bins. 

```python
# Especifico el número de bins
plt.hist(flowers_df.sepal_width, bins=5);
```

```python
import numpy as np

# Especifico los límites de cada bin
plt.hist(flowers_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
# Bins de distinto tamaño
plt.hist(flowers_df.sepal_width, bins=[1, 3, 4, 4.5]);
```

## Histogramas múltiples

Similar a los gráficos de lineas, podemos dibujar múltiples histogramas en un solo gráfico. Podemos reducir la opacidad de cada histograma, , de manera que un histograma no tape a los otros.

Vamos a dibujar histogramas separados para cada especie de flores


```python
setosa_df = flowers_df[flowers_df.species == 'setosa']
versicolor_df = flowers_df[flowers_df.species == 'versicolor']
virginica_df = flowers_df[flowers_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha=1, bins=np.arange(2, 5, 0.25));
plt.hist(versicolor_df.sepal_width, alpha=0.5, bins=np.arange(2, 5, 0.25));
```

También se pueden apilar varios instagramas uno arriba del otro.

## Gráficos de barras. 

Los gráficos de barras son similares a los de línea, por ejemplo, muestran una secuencia de valores. De todas maneras, se muestra una barra para cada valor en vez de ser puntos conectados por líneas. Podemos usar la funcion `plt.bar` para dibujar un gráfico de barra. 


```python
years = range(2000, 2006)
apples = [0.35, 0.6, 0.9, 0.8, 0.65, 0.8]
oranges = [0.4, 0.8, 0.9, 0.7, 0.6, 0.8]
```

```python
plt.bar(years, oranges);
```

Cómo los histogramas, podemos apilar las barras una sobre la otra. Para eso usamos el argumento `bottom`de `plt.bar`.

```python
plt.bar(years, apples)
plt.bar(years, oranges, bottom=apples);
```
